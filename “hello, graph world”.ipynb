```python
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
#----------initialising the graph
A = nx.Graph()

B = nx.generators.krackhardt_kite_graph()

print("Here are the nodes for the krackhardt_kite_graph: ","\n",B.nodes(),"\n")
print("Here are the edges for the krackhardt_kite_graph: ","\n",B.edges())


states =  {2:"Alabama" ,5:"Alaska", 7:"Arizona", 8:"Arkansas", 9:"California", 6:"New York", 4:"Florida", 1:"Hawaii", 3:"Ohio", 0:"Utah"}
states

#--------------Creating the graph object
A1 =nx.Graph()
#---------------Creating  Edges from the graph

Ed = [(0,1), (0,2), (0,6), (1,2), (1,3), (1,6), (2,4), (2,6), (2,7),
     (3,4), (3,6), (3,5), (4,6), (4,5), (4,7), (5,6), (7,8), (8,9)]


Ed = B.edges()

# Add all the edges
A1.add_edges_from(Ed)


C = nx.relabel_nodes(A1, states)

print("----------------------\n")

print(" Nodes for the graph: ", "\n", C.nodes())

print("----------------------\n")

print(" Edges for the graph: ", "\n", C.edges())

nx.draw(C, with_labels =1)
plt.show()
```
